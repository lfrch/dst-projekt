```python
import pandas as pd
import altair as alt

df = pd.read_csv("/Users/Lea/Desktop/dst-projekt/exitpoverty_fulldata.csv")
```

```python
df
```

```python
df.drop('FREQUENCY', axis=1, inplace=True)
```

```python
df.drop('Flag Codes', axis=1, inplace=True)
```

```python
df.drop('INDICATOR', axis=1, inplace=True)
```

```python
df4_selectedlocations = df[df['LOCATION'].isin(['OECD', 'EU27', 'DEU', 'GBR'])]
```

```python
df4_selectedlocations = df4_selectedlocations[df4_selectedlocations['SUBJECT'].isin(['SNGL2CHLD','SNGLNOCLD'])]
```

```python
df4_selectedlocations = df4_selectedlocations[df4_selectedlocations['MEASURE'].isin(['HR_67PC_AVEWAGE','HR_AVEWAGE'])]
```

```python
df4_selectedlocations
```

```python
df4_selectedlocations.rename(columns={'Value': 'VALUE'}, inplace=True)
```

```python
df4_selectedlocations.rename(columns={'MEASURE': 'WAGE'}, inplace=True)
```

```python
df_parttimewage = df4_selectedlocations[df4_selectedlocations['WAGE'] == 'HR_67PC_AVEWAGE']
df_parttimewage
```

```python
df_averagewage = df4_selectedlocations[df4_selectedlocations['WAGE'] == 'HR_AVEWAGE']
df_averagewage
```

```python
scatterplot = alt.Chart(df4_selectedlocations).mark_circle().encode(
    x=alt.X('VALUE', title='Value', scale=alt.Scale(zero=False)),
    y=alt.Y('VALUE', title='Value', scale=alt.Scale(zero=False)),
    color=alt.Color('WAGE', title='Wage'),
    tooltip=['SUBJECT', 'LOCATION', 'VALUE']
).transform_filter(
    alt.FieldOneOfPredicate(field='WAGE', oneOf=['HR_AVEWAGE', 'HR_67PC_AVEWAGE'])
)
scatterplot
```

```python
merged_df = pd.merge(df_parttimewage, df_averagewage, how='outer', on=['WAGE','VALUE'])

#merged_df = pd.concat([df_parttimewage, df_averagewage], axis=1, join="outer") #variable muss 1:1 identisch sein
merged_df

```

```python
scatterplot = alt.Chart(merged_df).mark_circle().encode(
    x='VALUE_x',
    y='VALUE_y',
    color='WAGE',
    #tooltip=['SUBJECT','LOCATION','annual_VALUE','weekly_VALUE']
)
scatterplot
```
